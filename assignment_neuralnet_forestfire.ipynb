```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('forestfires.csv')
```

```python
data.head()
```

```python
data.columns
```

```python
data=data.drop("monthfeb",axis=1)
data=data.drop("monthjan",axis=1)
data=data.drop("monthjul",axis=1)
data=data.drop("monthjun",axis=1)
data=data.drop("monthmar",axis=1)
data=data.drop("monthmay",axis=1)
data=data.drop("monthnov",axis=1)
data=data.drop("monthoct",axis=1)
data=data.drop("monthsep",axis=1)
data=data.drop("dayfri",axis=1)
data=data.drop("daymon",axis=1)
data=data.drop("daysat",axis=1)
data=data.drop("daythu",axis=1)
data=data.drop("daysun",axis=1)
data=data.drop("daytue",axis=1)
data=data.drop("daywed",axis=1)
data=data.drop("monthapr",axis=1)
data=data.drop("monthaug",axis=1)
data=data.drop("monthdec",axis=1)
```

```python
data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

data["month"]=le.fit_transform(data["month"])
data["day"]=le.fit_transform(data["day"])
data["size_category"]=le.fit_transform(data["size_category"])
data.head()
```

```python
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
seed = 7
numpy.random.seed(seed)


# split into input (X) and output (Y) variables
X = data[['month','day','FFMC','DMC','DC','ISI','temp','RH','wind','rain','area']]
Y = data["size_category"]
```

```python
model = Sequential()
model.add(Dense(12, input_dim=11, kernel_initializer='uniform', activation='relu'))
model.add(Dense(10, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adadelta', metrics=['accuracy'])
```

```python
model.fit(X, Y, validation_split=0.33, epochs=50, batch_size=10)
```

```python
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
model.history.history.keys()
```
